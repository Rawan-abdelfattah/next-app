```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.action_chains import ActionChains
import time
from selenium.webdriver.common.action_chains import ActionChains
import json
```

```python

chrome_service = ChromeService(executable_path='C:/Users/User/Downloads/webdriver/chromedriver.exe')  # Replace with the path to your chromedriver.exe
driver = webdriver.Chrome(service=chrome_service)

try:
    driver.get('https://symptoms.webmd.com/')

    # Find and click the button element
    button_element = driver.find_element(By.CSS_SELECTOR, 'button#male')  # Using the 'id' attribute
    button_element.click()

    # Find the input element by its 'id' attribute and enter "21"
    age_input = driver.find_element(By.CSS_SELECTOR, 'input#age')  # Using the 'id' attribute
    age_input.clear()  # Clear any existing text in the input
    age_input.send_keys("21")  # Enter "21" into the input

    # Sleep to give time for the input to update (you can replace this with a wait)
    time.sleep(2)
        # Find and click the "Continue" button element
    continue_button = driver.find_element(By.CSS_SELECTOR, 'button.continue-button')
    continue_button.click()
    time.sleep(2)
    
    path_element = driver.find_element(By.CSS_SELECTOR, 'path.layers')
    path_element.click()    
    time.sleep(2)
    
    back_button = driver.find_element(By.CSS_SELECTOR, 'div.body-symptoms-back-button')
    back_button.click()
    time.sleep(2)
    
    body_parts_element = driver.find_elements(By.CSS_SELECTOR, 'li.body-parts-list')[7]  
    body_parts_element.click()
    time.sleep(2)

    first_body_part = driver.find_element(By.XPATH, '//li[contains(text(), "Lower Abdomen")]')
    first_body_part.click()
    time.sleep(2)
    
    first_tab = driver.find_element(By.CSS_SELECTOR, '.body-symptoms-tabs .body-symptoms-tab:first-child')
    first_tab.click()
    time.sleep(2)
    
    
    # Find the <ul> element containing the <li> elements
    ul_element = driver.find_element(By.CSS_SELECTOR, 'div.body-symptoms-body ul')

    # Find all the <li> elements within the <ul>
    li_elements = ul_element.find_elements(By.CSS_SELECTOR, 'li.single-common-symptom')

    # Create a list to store the extracted data
    data = []

    # Iterate through the <li> elements and extract the text
    for li_element in li_elements:
        symptom_text = li_element.find_element(By.CSS_SELECTOR, 'span.body-symptom-text').text
        data.append(symptom_text)

    # Define the file path where you want to save the data
    file_path = 'D:/hospital/symptom-checker/src/data/lowerAbdomen.json'

    symptoms_dict = {'symptoms': data}

    # Create a dictionary to store the data
    symptoms_dict = {'symptoms': data}

    # Write the extracted data to a JSON file
    with open(file_path, 'w', encoding='utf-8') as json_file:
        json.dump(symptoms_dict, json_file, ensure_ascii=False, indent=4)

    
    
finally:
    # Quit the browser
    driver.quit()
```
